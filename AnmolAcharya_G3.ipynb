Task 4. Analyze Average Time Use:
Scenario: Each record contains daily hours of study, entertainment, and sleep.

2. Iterate over time_data and extract values into each list.
3. Calculate and print:

```python
study_list = []
ent_list = []
sleep_list = []

for day in time_data:
    study_list.append(day[0])
    ent_list.append(day[1])
    sleep_list.append(day[2])

def average(numbers):
    total = 0
    count = 0

    for n in numbers:
        total += n
        count += 1

    return total / count

print("Average study hours:", average(study_list))
print("Average entertainment hours:", average(ent_list))
print("Average sleep hours:", average(sleep_list))
```

```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

# Test
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(nested_list))
```

Task 2: Generate All Permutations of a String

```python
def generate_permutations(s):
    """Returns a sorted list of all possible rearrangements of the letters in s."""
    if len(s) <= 1:
        return [s]

    result = []
    for i in range(len(s)):
        char = s[i]
        rest = s[:i] + s[i+1:]
        for perm in generate_permutations(rest):
            result.append(char + perm)


    return sorted(list(set(result)))

print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

Task 3: Directory Size Calculation

```python
def calculate_directory_size(directory):
    """Adds up all file sizes in a folder, including files in subfolders."""
    total = 0

    for name, content in directory.items():
        if type(content) == dict:
            total += calculate_directory_size(content)
        else:
            total += content

    return total


folder = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print("Total size:", calculate_directory_size(folder), "KB")
```

 Task 1: Coin Change (Min Coins)

```python
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1,2,5], 11))
```

Task 2: Longest Common Subsequence (LCS)

```python
def longest_common_subsequence(s1, s2):
    n, m = len(s1), len(s2)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(1, n+1):
        for j in range(1, m+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[n][m]

print(longest_common_subsequence("abcde", "ace"))
```

Task 3 - 0/1 Knapsack Problem:
1. Write a function knapsack(weights, values, capacity) that:
• Uses DP to determine the maximum value that can be achieved within the given
weight capacity.

2. Test with weights [1, 3, 4, 5], values [1, 4, 5, 7], and capacity 7. The re-
sult should be 9.

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [0] * (capacity + 1)

    for i in range(n):
        for w in range(capacity, weights[i] - 1, -1):

            dp[w] = max(dp[w], dp[w - weights[i]] + values[i])

    return dp[capacity]
weights = [1, 3, 4, 5]
values  = [1, 4, 5, 7]
capacity = 7

print("Maximum value:", knapsack(weights, values, capacity))
```
